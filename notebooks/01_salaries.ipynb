```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carica il dataset
df = pd.read_csv('../data/ds_salaries.csv')

# Prima occhiata ai dati
print(df.shape)
df.head()
```

```python
# Stipendi medi per job title (top 10)
plt.figure(figsize=(12, 6))
top_jobs = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=top_jobs.values, y=top_jobs.index, palette='viridis')
plt.title('Top 10 Job Titles by Average Salary (USD)')
plt.xlabel('Average Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
# Evoluzione stipendi medi per anno
plt.figure(figsize=(10, 5))
yearly = df.groupby('work_year')['salary_in_usd'].mean()
sns.lineplot(x=yearly.index, y=yearly.values, marker='o', color='teal')
plt.title('Average Data Science Salary Over Time')
plt.xlabel('Year')
plt.ylabel('Average Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
# Stipendi per experience level
plt.figure(figsize=(10, 5))
exp_order = ['EN', 'MI', 'SE', 'EX']
exp_labels = {'EN': 'Entry', 'MI': 'Mid', 'SE': 'Senior', 'EX': 'Executive'}
df['exp_label'] = df['experience_level'].map(exp_labels)
sns.boxplot(x='exp_label', y='salary_in_usd', data=df, 
            order=['Entry', 'Mid', 'Senior', 'Executive'], palette='coolwarm')
plt.title('Salary Distribution by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.tight_layout()
plt.show()
```
